.זה מחשב לפי הסימן 
מה קוד הבא אמור לעושה?

```python
def do_math(a, math, b):
    if math == "+": return a + b
    if math == "-": return a - b
    if math == "*": return a * b
    if math == "/": return a / b
    if math == "//" : return a // b  
    if math == "%" : return a % b
    return "I don't know what is: " + math
```

```python
print(do_math(10, "+", 20))
print(do_math(10, "-", 20))
print(do_math(10, "*", 20))
print(do_math(10, "/", 20))
```

--- הקוד מחלק את המספר ומדפיס את השארית
מה קוד הבא עושה?
<br>
איך הוא עושה את זה?

```python
def printDigits(num):
    while (num > 0):
        print(num % 10)
        num = num // 10
```

```python
printDigits(9876450)
```

--- סוכם את כל הספרות שמרכיבות את המספר
מה קוד הבא עושה?
<br>
איך הוא עושה את זה?

```python
def secret(num):
    sod = 0
    while (num > 0):
        sod = sod + num % 10
        num = num // 10
    return sod
```

```python
print(secret(123456789))
```

```python
print(secret(123456789))
print(secret(secret(123456789)))
print(secret(secret(secret(123456789))))
```
