מגדיר סימנים מתמטיים ומחשב אותם 

```python
def do_math(a, math, b):
    if math == "+": return a + b
    if math == "-": return a - b
    if math == "*": return a * b
    if math == "/": return a / b
    if math == "//": return a // b
    if math == "%": return a % b
    return "I don't know what is: " + math
```

```python
print(do_math(10, "+", 20))
print(do_math(10, "-", 20))
print(do_math(10, "*", 20))
print(do_math(10, "/", 20))
```

הוא מדפיס מהסוף להתחלה בכך שהוא מחלק ולוקח את השארית כל פעם עד שמגיע למספר 0 או מתחת לו 

```python
def printDigits(num):
    while (num > 0):
        print(num % 10)
        num = num // 10
```

```python
printDigits(12365)
```

מחבר את כל המספרים יחד  על ידי הפרדת ספרות  על ידי חילוק ולקיחת שארית


```python
def secret(num):
    sod = 0
    while (num > 0):
        sod = sod + num % 10
        num = num // 10
    return sod
```

```python
print(secret(123456789))
```

```python
print(secret(123456789))
print(secret(secret(123456789)))
print(secret(secret(secret(123456789))))
```
