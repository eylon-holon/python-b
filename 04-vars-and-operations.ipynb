מה קוד הבא אמור לעושה?

```python
def do_math(a, math, b):
    if math == "+": return a + b
    if math == "-": return a - b
    if math == "*": return a * b
    if math == "/": return a / b
    if math == "//": return a // b
    if math == "%": return a % b
    return "I don't know what is: " + math
```

```python
print(do_math(10, "+", 20))
print(do_math(10, "-", 20))
print(do_math(10, "*", 20))
print(do_math(10, "/", 20))
```

הקוד מפרק מספרים לספרות המרכיבות אותן
והוא מחלק את המספר וכותב את השארית שלו
לאחר מכן הוא מחלק את המספר המקורי שוב ב10 ושומר אותו.
התהליך נמשך עד שיוצאת לנו תוצאה שהיא שווה ל0

---
מה קוד הבא עושה?
<br>
איך הוא עושה את זה?

```python
def secret(num):
    sod = 0
    while (num > 0):
        sod = sod + num % 10
        num = num // 10
    return sod
```

```python
print(secret(123456789))
```

הקוד סוכם את כל הספרות שמרכיבות את המספר שבחרנו

```python
print(secret(123456789))
print(secret(secret(123456789)))
print(secret(secret(secret(123456789))))
```
