מגדיר את הפונקציה ואת הפעולות חשבון
---
מה קוד הבא אמור לעושה?

```python
def do_math(a, math, b):
    if math == "+": return a + b
    if math == "-": return a - b
    if math == "*": return a * b
    if math == "/": return a / b
    if math == "%": return a % b
    if math == "//": return a // b
    return "I dont know what it is: " + math
```

```python
print(do_math(10, "+", 20))
print(do_math(10, "-", 20))
print(do_math(10, "*", 20))
print(do_math(10, "/", 20))
```

קודם כל מגדיר פונקציה ואז הוא נכנס ללולאה כי נאם גדול מ0 ואז עושה חילוק שארית של 10 וחילוק שלמים של 10 
----
מה קוד הבא עושה?
<br>
איך הוא עושה את זה?

```python
def printDigits(num):
    while (num > 0):
        print(num % 10)
        num = num // 10
```

```python
printDigits(12345)
```

מחבר את הספרות
 וכשיש ספרה אחת אז היא תשאר אותו דבר לא משנה כמה סיקרט יש
 ---
מה קוד הבא עושה?
<br>
איך הוא עושה את זה?

```python
def secret(num):
    sod = 0
    while (num > 0):
        sod = sod + num % 10
        num = num // 10
    return sod
```

```python
print(secret(123456789))
```

```python
print(secret(123456789))
print(secret(secret(123456789)))
print(secret(secret(secret(123456789))))
```
