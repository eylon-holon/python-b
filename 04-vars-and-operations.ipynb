```python
def secret(num):
    sod = 0
    while (num > 0):
        sod = sod + num % 10
        num = num // 10
    return sod
```

```python
print(secret(123456789))
```

```python
print(secret(123456789))
print(secret(secret(123456789)))
print(secret(secret(secret(123456789))))
```
