```python
def findMin(a):
    min = a[0]
    for i in range(0, len(a), 1):
        if a[i] < min:
            min = a[i]
    return min
```

```python
print(findMin([10, 20, 30, -10, 10]))
print(findMin([10, 20, -30, -10, 10]))
print(findMin([10, -20, 30, -10, 10]))
print(findMin([10, 20, 30, 10, 10]))
```
