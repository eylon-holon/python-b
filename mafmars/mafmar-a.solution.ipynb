````python
first = 1
second = 1
counter = 3
print("Fibonacci item 1 is ", first)
print("Fibonacci item 2 is ", second)
while counter <= 8:
    third = first + second
    first = second
    second = third
    print("Fibonacci item", counter, "is ", third)
    counter = counter + 1
````
