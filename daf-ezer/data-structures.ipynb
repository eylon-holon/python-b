### <span style="color:magenta">Queue - תור

|פעולה      | מה עושה                                                  |que | x |
|---                        |---                                        |---        |--- |
|<tt>import Queue           | שורה זו חייבת להיות  בכל תוכנית המשתמשת בתור |<tt>       | |
|<tt>que = Queue.queue()    | יוצר תור חדש ריק                  |<tt>[]         | |
|<tt>que.put(10)            | להוסיף מספר לתור                  |<tt>[10]       | |
|<tt>que.put(20)            | מספר תמיד מתווסף לסוף תור        |<tt>[10,20]    | |
|<tt>x = que.get()          | להוציא מספר מתור ולאחזיר אותו    |<tt>[20]       |<tt>10 |
|<tt>x = que.get()          |                                    |<tt>[]         |<tt>20 |


````python
def last(q1):
    q1.put(-999)
    ret_val = -1
    temp = q1.get()
    while temp != -999:
        ret_val = temp
        q1.put(temp)
        temp = q1.get()
    return ret_val
````

| פקודה                             |while      | q1               | ret_val      | temp   |
|---                                |---        |---                |---        |---        |

|<tt>temp = q1.get()                |           |<tt>[7,10,3,-999]  |           |<tt>4      |

|<tt>&nbsp; &nbsp; q1.put(temp)     |           |<tt>[7,10,3,-999,4]|           |           |
|<tt>&nbsp; &nbsp; temp = q1.get()  |           |<tt>[10,3,-999,4]  |           |<tt>7      |
|<tt>while temp != -999             |<tt>True   |                   |           |           | 
|<tt>&nbsp; &nbsp; ret_val = temp   |           |                   |<tt>7      |           |
|<tt>&nbsp; &nbsp; q1.put(temp)     |           |<tt>[10,3,-999,4,7]|           |           |
|<tt>&nbsp; &nbsp; temp = q1.get()  |           |<tt>[3,-999,4,7]   |           |<tt>10     |

````python
import queue

q1 = queue.Queue()
q1.put(4)
q1.put(7)
q1.put(10)
q1.put(3)
temp = last(q1)
print(temp)
````

### <span style="color:magenta"> Tuple

|Tuple      | הסבר        |תוצאה |
|---       |---                 |--- |
|<tt>t = ('Avi', 91)           | זוג של ערכים |<tt>       |
|<tt>print(t)           |  |<tt>('Avi', 91)       |
|<tt>print("name:", t[0], "grade:", t[1])           | גישה לאיברים דומה לגישה ברשימה |<tt>name: Avi grade: 91       |
|<tt>t[0] = "Beni"  | tuple אי אפשר לשנות ערך של |<tt> Error       |
|<tt>a = [t, ('Beni', 89)] | tupes רשימה של |<tt>[('Avi',91), ('Beni',89)]       |
|<tt>print("name:", a[1][0], "grade:", a[1][1])           |  |<tt>name: Beni grade: 89       |
|<tt>a2 = [(1,5), (3,7), (2,10)] | כאשר כל אחד זה זוג של מספרים tupes רשימה של |<tt>       |



````python
t = ("Avi", 85)
print(t)
print(t[0], t[1])
students = [t, ("Beni",90), ("Gili",95)]
for stu in students:
    if (stu[1] > 90):       # הדפס את אלה שציונם גדול מ-90
        print("name: ", stu[0], ", grade: ", stu[1])
````
