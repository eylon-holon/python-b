```python
def what_02(a, b, c):
    if a + b < c:
        return False
    if a + c < b:
        return False
    if b + c < a:
        return False
    return True
```

```python
def what_03(a):
    what = a[0]
    for i in range(0, len(a), 1):
        if a[i] < what:
            what = a[i]
    return what
```

check if the number in line a and b are the same

```python
def what_06(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return True
    return False
```

```python
print(what_06([5,1,3,4,2], 5))
```

check if the numbers in line a equal to num

```python
def what_07(a):
    what = 0
    for i in range(len(a)):
        what = what + a[i]
    return what
```

```python
def what_08(a, b):
    for i in range(len(b)):
        if a[len(a)-i-1] != b[len(b)-i-1]:
            return False
    return True
```

```python
print(what_08([5,1,3,4,2], [5,1,3,4,2]))
```

check if two lines are the same

```python
def what_09(a, num):
    for i in range(len(a)):
        if a[i] != num:
            return False
    return True
```

```python
print(what_09([1,2,3,4,5], 10))
print(what_09([5,1,3,4,2], 5))
print(what_09([5,5,5,5], 5))
```
