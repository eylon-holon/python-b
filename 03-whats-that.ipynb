```python
def what_01(a, b):
    if a < b:
        return a
    else:
        return b
```

print(what_01(5, 10))  # Output: 5
print(what_01(10, 5))  # Output: 5
print(what_01(7, 7))   # Output: 7


```python
def what_02(a, b, c):
    if a + b < c:
        return False
    if a + c < b:
        return False
    if b + c < a:
        return False
    return True
```

```python
def what_04(num):
    return num % 2 == 0
```

```python
def what_05(a, b):
    for i in range(0, len(b), 1):
        if a[i] != b[i]:
            return False
    return True
```

print(what_05([1, 2, 3], [1, 2, 3]))  # Output: True
print(what_05([1, 2, 3], [3, 2, 1]))  # Output: False
print(what_05([1, 2], [1, 2, 3]))    # Output: False (different lengths)


```python
def what_06(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return True
    return False
```

print(what_06([1, 2, 3, 4, 5], 3))  # Output: True
print(what_06([1, 2, 3, 4, 5], 6))  # Output: False
print(what_06([], 3))               # Output: False

```python
def what_08(a, b):
    for i in range(len(b)):
        if a[len(a)-i-1] != b[len(b)-i-1]:
            return False
    return True
```

print(what_08([1, 2, 3], [3, 2, 1]))  # Output: True (reverse of each other)
print(what_08([1, 2, 3], [3, 1, 2]))  # Output: False (not reverse)
print(what_08([1, 2], [1, 2, 3]))    # Output: False (different lengths)


```python
def what_09(a, num):
    for i in range(len(a)):
        if a[i] != num:
            return False
    return True
```

print(what_09([3, 3, 3], 3))  # Output: True (all elements are 3)
print(what_09([1, 2, 3], 3))  # Output: False (not all elements are 3)
print(what_09([7, 7, 7], 7))  # Output: True (all elements are 7)


```python
def what_10(a):
    return what_09(a, a[0])
```

```python
def what_11(a, b, c):
    return a*a + b*b == c*c
```

print(what_11(3, 4, 5))  # Output: True (3^2 + 4^2 = 5^2)
print(what_11(5, 12, 13))  # Output: True (5^2 + 12^2 = 13^2)
print(what_11(1, 2, 3))  # Output: False (1^2 + 2^2 != 3^2)



```python
def what_12(a, b):
    return a % b == 0
```

print(what_12(10, 2))  # Output: True (10 is divisible by 2)
print(what_12(10, 3))  # Output: False (10 is not divisible by 3)
print(what_12(15, 5))  # Output: True (15 is divisible by 5)
print(what_12(7, 3))   # Output: False (7 is not divisible by 3)


```python
def what_13(num):
    for i in range(2, num-1, 1):
        if what_12(num, i):
            return False
    return True
```

print(what_13(2))  # Output: True (2 is prime)
print(what_13(3))  # Output: True (3 is prime)
print(what_13(4))  # Output: False (4 is not prime)
print(what_13(17)) # Output: True (17 is prime)
print(what_13(18)) # Output: False (18 is not prime)
