```python
def what_01(a, b):
    if a < b:
        return a
    else:
        return b
```

הפונקציה תחזיר את המספר הקטן יותר 

```python
def what_02(a, b, c):
    if a + b < c:
        return False
    if a + c < b:
        return False
    if b + c < a:
        return False
    return True
```

אם יהיה אפשר להרכיב רכיב משולש ממספרים הללו 

```python
def what_03(a):
    what = a[0]
    for i in range(0, len(a), 1):
        if a[i] < what:
            what = a[i]
    return what
```

בודק מה המספר הכי גדול

```python
def what_04(num):
    return num % 2 == 0
```

האם המספר מתחלק בשתיים והוא זוגי

```python
def what_05(a, b):
    for i in range(0, len(b), 1):
        if a[i] != b[i]:
            return False
    return True
```

בודק את האורך 

```python
def what_06(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return True
    return False
```

בודק האם המספר נמצא במחרוזת

```python
def what_07(a):
    what = 0
    for i in range(len(a)):
        what = what + a[i]
    return what
```

בודק האם המספרים בלולאה ב שווים להתחלה של לולאה א

```python
def what_08(a, b):
    for i in range(len(b)):
        if a[len(a)-i-1] != b[len(b)-i-1]:
            return False
    return True
```

בודק האם המספר שונה מהמערך

```python
def what_09(a, num):
    for i in range(len(a)):
        if a[i] != num:
            return False
    return True
```

אם יש את אותו מספר בכל המערך

```python
def what_10(a):
    return what_09(a, a[0])
```

פיתגורס

```python
def what_11(a, b, c):
    return a*a + b*b == c*c
```

אם השארית של שני המספרים שווה 0 כלומר אם המספרים מתחלקים זה בזה

```python
def what_12(a, b):
    return a % b == 0
```

בודק האם יש מספריים ראשונים ומחלק אותו בכל המספרים עד אותו מספר ואם ה לא מתחלק באף אחד זה נכון

```python
def what_13(num):
    for i in range(2, num-1, 1):
        if what_12(num, i):
            return False
    return True
```
