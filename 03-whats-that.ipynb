```python
def what_01(a, b):
    if a < b:
        return a
    else:
        return b
```

מחזירה את מה שיותר קטן

```python
def what_02(a, b, c):
    if a + b < c:
        return False
    if a + c < b:
        return False
    if b + c < a:
        return False
    return True
```

בודקת אם אפשר להרקיב משולש

```python
def what_03(a):
    what = a[0]
    for i in range(0, len(a), 1):
        if a[i] < what:
            what = a[i]
    return what
```

מקבל מספר ומחזיר אותו אם הוא מתחלק בשתיים ללא שארית

```python
def what_05(a, b):
    for i in range(0, len(b), 1):
        if a[i] != b[i]:
            return False
    return True
```

מקבל שני מספרים ויוצר מחרוזרת באורך של
B בקפיצןת של 1 ואם המיקום שךA לא שווה למיקום של B הוא מחזיר FALSE ואם הוא כן הוא מחזיר TRUE

מקבל שני מספרים ובודק אם A במיקום של NUM ואם כן מחזיר TRUE ןאם לא הוא מחזיר 

```python
def what_07(a):
    what = 0
    for i in range(len(a)):
        what = what + a[i]
    return what
print (what_07([10,20,30]))
```

עושה סדרה של מספרים ומוסיף לWHAT את כל האיברים ומחזיר את WHAא עם סכום המספרים 

```python
def what_08(a, b):
    for i in range(len(b)):
        if a[len(a)-i-1] != b[len(b)-i-1]:
            return False
    return True
```

עושה סדרה של מספרים ובודק אם A פחות I פחות 1 לא שווה לB פחות I פחות 1 ןאם זה לא שווה הוא מחזיר FALSE
ואם זה כן שווה זה מחזיר TRUE

```python
def what_09(a, num):
    for i in range(len(a)):
        if a[i] != num:
            return False
    return True
```

עטשה סדרה שך מספרים ואם A לא שווה ל NUM הוא מחזיר FALSE ואם זה כן שווה הוא מחזיר TRUE

```python
def what_10(a):
    return what_09(a, a[0])
```

מחזיר TRUE או FALSE אם המספרים שהוצבו מתאימים 

```python
def what_12(a, b):
    return a % b == 0
```

מחזיר TRUE או FALSE אם השארית של המספרים שהוצבו שווה ל0

```python
def what_13(num):
    for i in range(2, num-1, 1):
        if what_12(num, i):
            return False
    return True
```
