```python
def what_01(a, b):
    if a < b:
        return a
    else:
        return b
```

מראה את המספר הנמוך יותר

```python
def what_02(a, b, c):
    if a + b < c:
        return False
    if a + c < b:
        return False
    if b + c < a:
        return False
    return True
```

האם אפשר לרכיב משולש מהצלעות הללו

```python
def what_03(a):
    what = a[0]
    for i in range(0, len(a), 1):
        if a[i] < what:
            what = a[i]
    return what
```

מה המספר הכי קטן במחרוזת

```python
def what_04(num):
    return num % 2 == 0
```

האם המספר הוא זוגי

```python
def what_05(a, b):
    for i in range(0, len(b), 1):
        if a[i] != b[i]:
            return False
    return True
```

האם הלולאה השנייה היא ההתחלה של הראשונה

```python
def what_06(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return True
    return False
```

מוצא מספר מסויים במערך

```python
def what_07(a):
    what = 0
    for i in range(len(a)):
        what = what + a[i]
    return what
```

סכום המספרים במערך

```python
def what_08(a, b):
    for i in range(len(b)):
        if a[len(a)-i-1] != b[len(b)-i-1]:
            return False
    return True
```

  בודק האם המספרים במארך ב שווים לסוף של מארך א

```python
def what_09(a, num):
    for i in range(len(a)):
        if a[i] != num:
            return False
    return True
```

```python
print(what_09([1,2,3],1))
print(what_09([1,1,1],1))
```

בודק האם המערך מורכב מאותו המספר שהוא קיבל

```python
def what_10(a):
    return what_09(a, a[0])
```

```python
print(what_10([6,6,6,7]))
print(what_10([1,1,1]))
```

מוצא האם יש את אותו מספר כל המערך

```python
def what_11(a, b, c):
    return a*a + b*b == c*c
```

פיתגורס- האם המספרים הינם יכולים ליצור משולש ישר זווית

```python
def what_12(a, b):
    return a % b == 0
```

האם המספרים מתחלקים זה 

```python
def what_13(num):
    for i in range(2, num-1, 1):
        if what_12(num, i):
            return False
    return True
```

```python
print(what_13(11))
print(what_13(13))
print(what_13(12))
```

בודק האם זה מספר ראשוני

```python
def what_14(a):
    for i in range (1, len(a), 1):
        if a[i-1] > a[i]:
            return False
    return True
```

```python
print(what_14([1,2,3]))
print(what_14([3,2,1]))
```

מוצא אם המערך בסדר עולה

```python
def what_15(a):
    for i in range(len(a)-1, 0, -1):
        if a[i] < a[i-1]:
            return False
    return True
```

מוצא האם המערך בסדר עולה

```python
def what_16(a, a0, d):
    for i in range(len(a)):
        if a[i] != a0 + d * i:
            return False
    return True
```

```python
print(what_16([10,20,30],10,10))
print(what_16([1,2,3],1,1))
print(what_16([1,5,3],2,-2))
```

האם זו סידרה חשבונית

```python
def what_17(a):
    for i in range(0, len(a), 1):
        if a[i] != a[len(a)-1-i]:
            return False
    return True
```

```python
print(what_17([5,10,15,20,]))
print(what_17([20,10,7,10,20,]))
```
