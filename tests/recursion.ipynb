```python
def findMax(a):
    if len(a) == 1:
        return a[0]
    max = findMax(a[1::])
    if a[0] > max:
        return a[0]
    else:
        return max
```

```python
print(findMax([5,10,3,4,6]))
```

```python
def toBinaryString(num):
    s = ""
    while num > 0:
        s = str(num % 2) + s
        num = num // 2
    return s
```

```python
print(toBinaryString(75))
```

```python
def toOctalString(num):
    s = ""
    while num > 0:
        s = str(num % 8) + s
        num = num // 8
    return s
```

```python
print(toOctalString(75))
```

```python
def what(num):
    if num == 0:
        return ""
    return what(num // 2) + str(num % 2)
```

```python
print(what(100))
```
