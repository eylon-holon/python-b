```python
names=["Avi Avior", "Beni Benzi", "Gili Tet"]
count1 = 0
count2 = 0
count3 = 0
for name in names:
    if 'vi' in name:
        count1 += 1
    if len(name) > 8:
        count2 += 1
    parts = name.split()
    if parts[0][0] == parts[1][0]:
        count3 += 1
print("vi in names: ", count1)
print("long names: ", count1)
print("same letters: ", count1)
```
