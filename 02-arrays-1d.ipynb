=== M I N ===

```python
# return minimal value of an array
def min(a):
    m = a[0]
    for n in a:
        if n < m:
            m = n
    return m
```

```python
print(min([1, 2, 3]))
print(min([1, 2, -3]))
print(min([-1, -20, -3]))
```

```python
submit(min)
```

=== M A X ===

```python
# return maximal value of an array
def max(a):
    m=a[0]
    for n in a:
        if n>m:
            m=n
    return m
```

```python
print(max([1, 2, 3]))
print(max([1, 2, -3]))
print(max([-1, -20, -3]))
```

```python
submit(max)
```

=== F I N D ===

```python
# find first occurrence of n in array
# if found returns element's index
# if not found returns -1
def find(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return i
    return -1
```

```python
print(find([1, 2, 3], 2));
print(find([3, 1, 2], 3));
print(find([1, 3, 4], 2));
```

```python
submit(find)
```

=== F I N D - L A S T ===

```python
# finds last occurrence of n in array
# if found returns element's index
# if not found returns -1
def findLast(a, num):
    for i in range(len(a)-1,-1,-1):
       if a[i]==num:
           return i
    return -1
```

```python
print(findLast([1, 2, 3], 2))
print(findLast([1, 2, 2], 2))
print(findLast([1, 3, 4], 2))
```

```python
submit(findLast)
```

=== I S - P A L I N D R O M E ===

```python
# Function returns True if input string is palidrome
# and False if not
def isPalindrome(a):
     for i in range(len(a)//2):
          if a[i]==a[len(a)-i-1]:
               return True
     return False
```

```python
# WRITE TESTS FOR PALINDROME FUNCTION
print(isPalindrome([20,10,5,5,10,20]))
print(isPalindrome([20,12,10,4,2]))
print(isPalindrome([20,10,4,1,4,10,20]))
```

```python
submit(isPalindrome)
```

=== S T A R T S - W I T H ===

```python
# Check if array starts with another array - prefix
def startsWith(a, prefix):
    for i in range(0,len(prefix),1):
        if prefix[i]!=a[i]:
            return False
    return True
```

```python
# WRITE TESTS FOR startsWith
print(startsWith([10,40,60],[10,40]))
print(startsWith([10,50,60],[10,40]))
```

```python
submit(startsWith)
```

=== E N D S - W I T H ===

```python
# Check if array ends with another array - suffix
def endsWith(a, suffix):
    for i in range(0,len(suffix),1):
        if a[len(a)-i-1]!=suffix[len(suffix)-i-1]:
            return False
    return True
```

```python
# WRITE TESTS FOR endsWith
print(endsWith([10,20,30,40],[30,40]))
print(endsWith([10,20,30,40],[70,50]))
print(endsWith([10,20,30,40],[30,60]))
```

```python
submit(endsWith)
```
