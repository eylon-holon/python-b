=== M I N ===

```python
# return minimal value of an array
def max(a):
    max = a[0]
    for num in a:
        if num > max:
            max = num
    return max
```

```python
print(max([1, 2, 3]))
print(max([1, 2, -3]))
print(max([-1, -20, -3]))
```

```python
submit(max)
```

=== F I N D ===

```python
# find first occurrence of n in array
# if found returns element's index
# if not found returns -1
def find(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return i
    return -1
```

```python
print(find([2, 1, 2, 3], 2));
print(find([3, 1, 2], 3));
print(find([1, 3, 4, 10], 2));
```

```python
submit(find)
```

=== F I N D - L A S T ===

```python
# finds last occurrence of n in array
# if found returns element's index
# if not found returns -1
def findLast(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return i
    return -1
```

```python
print(findLast([1, 2, 3, 2], 2))
print(findLast([1, 2, 2], 2))
print(findLast([1, 2, 3, 4, 2], 2))
```

```python
submit(findLast)
```

=== I S - P A L I N D R O M E ===

```python
# Function returns True if input string is palidrome
# and False if not
def isPalidrome(a):
    for i in range(0,len(a),1):
        if a[i] != a[len(a)-i-1]:
            return False
        else:
            return True
```

```python
# WRITE TESTS FOR PALINDROME FUNCTION
print(isPalindrome([1,2,3,2,1]))
print(isPalindrome([1,3,5,4,6]))
print(isPalindrome([10,20,88,20,10]))
```

```python
submit(isPalidrome)
```

=== S T A R T S - W I T H ===

```python
# Check if array starts with another array - prefix
def startsWith(a, prefix):
    for i in range(0,len(prefix),1):
        if prefix[i] != a[i]:
            return False
    return True
```

```python
# WRITE TESTS FOR startsWith
print(startsWith([1,2,3,2,1],[1,2,3]))
print(startsWith([1,2,3,2,1],[1,2,4]))
print(startsWith([1,5,2,2,1],[1,5,2]))
```

```python
submit(startsWith)
```

=== E N D S - W I T H ===

```python
# Check if array ends with another array - suffix
def endsWith(a, suffix):
    for i in range(0,len(suffix),1):
        if suffix[len(suffix)-i-1] != a[len(a)-i-1]:
            return False
    return True
```

```python
# WRITE TESTS FOR endsWith
print(endsWith([1,2,3,2,1],[3,2,1]))
print(endsWith([1,5,3,5,1],[3,2,1]))
print(endsWith([1,2,5,10,1],[5,10,1]))
```

```python
submit(endsWith)
```
