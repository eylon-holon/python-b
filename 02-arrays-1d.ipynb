=== M I N ===

```python
# return minimal value of an array
def min(a):
    m = a[0]
    for n in a:
        if n < m:
            m = n
    return m
```

```python
print(min([1, 2, 3]))
print(min([1, 2, -3]))
print(min([-1, -20, -3]))
```

```python
submit(min)
```

=== M A X ===

```python
# return maximal value of an array
def max(a):
    max= a[0]
    for num in a:
        if num > max:
            max = num
    return max
```

```python
print(max([1, 4, 3]))
print(max([1, 2, -3]))
print(max([-1, -20,123, -3]))
```

```python
submit(max)
```

=== F I N D ===

```python
# find first occurrence of n in array
# if found returns element's index
# if not found returns -1
def find(a, num):
    for i in range(len(a)):
        if a[i] == num:
            return i
    return -1
```

```python
print(find([2, 10, 8,5,1,10], 10));
 
print(find([3, 1, 2], 3));

print(find([1, 3, 4], 2));
         
```

```python
submit(find)
```

=== F I N D - L A S T ===

```python
# finds last occurrence of n in array
# if found returns element's index
# if not found returns -1
def findLast(a, num):
    for i in  range(len(a)-1, -1, -1):
        if a[i] == num:
            return i
    return -1
```

```python
print(findLast([2, 3, 3], 2))
print(findLast([1, 2, 2, 4, 7, 2], 2))
print(findLast([1, 3, 4], 2))
```

```python
submit(findLast)
```
