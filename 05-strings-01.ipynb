count - מה הפונקציה מחזירה?

```python
str = "0123456789ABCDEF"
print(str.find("0"))
print(str.find("01"))
print(str.find("789"))
print(str.find("EF"))
print(str.find("FE"))
```

replace - מה הפונקציה מחזירה?

היא מאתרת את הרצף מילים שנתת לו ומחליף אותו במשהו אחר

```python
print("1234".isnumeric())
print("1234".isalpha())
print("abcd".isnumeric())
print("abcd".isalpha())
```

isnumeric(), isalpha()<br>
מה הפונקציות מחזירות?

isnumeric- מראה האם הרצף מורכב מספרות בלבד
isalpha- מראה האם הרצף מורכב מאותיות בלבד

```python
print("-1234".isnumeric())
print("-1234".isalpha())
```

false - למה שניהם החזירו?

מינוס נחשב כאות לפיתון

```python
print("abcdef".upper())
print("ABCDEF".lower())
```

upper(), lower()<br>
מה הפונקציות מחזירות?

upper- הופך את כל הרצף לאותיוץת גדולות
lower-הופך את כל הרצף לאותיות קטנות 

```python
s1 = "One two THree"
s2 = "one TWO three"
print(s1.lower() == s2.lower())
print(s1.upper() == s2.upper())
```
