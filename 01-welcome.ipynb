תתחילו עבודה בלהריץ את המשבצת הבא

```python
from eylon.core import *
start({
    "hello()": [
        [[], [], ['Gosha'], ['Hi Gosha']]
    ],
    "min()": [
        [[10, 20], [10], [], []],
        [[20, 10], [10], [], []],
        [[-10, -20], [-20], [], []],
    ],
    "max()": [
        [[10, 20], [20], [], []],
        [[20, 10], [20], [], []],
        [[-10, -20], [-10], [], []],
    ],
})
```

עכשיו מימשו את הפונקציה

```python
def hello():
    name = input("Please, enter your name: ")
    print("Hi " + name)
```

ממשו פונקציה
min of two variables

```python
def secret(a, b):
    if a < b:
        return a
    else:
        return b  

def max(a, b):
    if a > b:
        return a
    else:
        return b
```

בידקו את הפונקציה

```python
print(secret(10, 20))
print(secret(20, 10))
print(secret(-10, 20))
print(secret(10, -20))
```

```python
print(max(10, 20))
print(max(20, 10))
print(max(-10, 20))
print(max(10, -20))
```

הגישו את הפונקציה

```python
submit(min)
```
