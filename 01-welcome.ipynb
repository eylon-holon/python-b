min of two variables

```python
def min(a, b):
    return a if a < b else b

def max(a, b):
    return a if a > b else b
```

拽 转 驻拽爪

```python
print(min(10, 20))
print(min(20, 10))
print(min(-10, 20))
print(min(10, -20))
```

```python
submit(min)
```
