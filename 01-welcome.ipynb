ממשו פונקציה
min of two variables

```python
def min(a, b):
    return a if a < b else b

def max(a, b):
    return a if a > b else b
```

בידקו את הפונקציה

```python
print(min(10, 20))
print(min(20, 10))
print(min(-10, 20))
print(min(10, -20))
```

הגישו את הפונקציה

```python
submit(min)
```
